```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/heart_failure_clinical_records_dataset.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
disease=['creatinine_phosphokinase','ejection_fraction','serum_sodium','time','platelets']
data[disease]=sc.fit_transform(data[disease])
```

```python
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
dis=['age','anaemia','diabetes','high_blood_pressure','serum_creatinine','sex','smoking','DEATH_EVENT']
data[dis]=mm.fit_transform(data[dis])
```

```python
data.head()
```

```python
x=data.iloc[:,:-1]
x
```

```python
y=data.iloc[:,-1:]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
data.head()
```

```python
data.corr()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.coef_,lr.intercept_
```

```python
y_pred=lr.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
rms=np.sqrt(mean_squared_error(y_test,y_pred))
rms
```

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/heart_failure_clinical_records_dataset.csv")
data
```

```python
x=data.iloc[:,:-1]
x
```

```python
y=data.iloc[:,-1:]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
model=TPOTClassifier(generations=5,population_size=20,verbosity=2,random_state=1502)
model.fit(x_train,y_train)
```

```python
accuracy=model.score(x_test,y_test)
accuracy
```
